```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Muralish/Desktop/Test/dummy_churn_data.csv")

X = df[["total_day_charge","total_eve_charge"]].values
y= df["churn"].values

X_train,X_test, y_train,y_test = train_test_split(X,y,test_size = .2, random_state = 42,stratify = y)

train_accuracies = {}
test_accuracies = {}
neighbors = np.arange(1,10)

for neighbor in neighbors:
    knn = KNeighborsClassifier(n_neighbors = neighbor)
    knn.fit(X_train,y_train)
    train_accuracies[neighbor] = knn.score(X_train,y_train)
    test_accuracies[neighbor] = knn.score(X_test,y_test)
    

plt.figure(figsize =(8,6))
plt.title("Finding the best K value")
plt.plot(neighbors,train_accuracies.values(),label="Train Accuracy")
plt.plot(neighbors,test_accuracies.values(),label="Test Accuracy")
plt.legend()
plt.xlabel("Number of Neighbors")
plt.ylabel("Accuracy")
plt.show()

```

```python
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train,y_train)
train_accuracy = knn.score(X_train,y_train)
test_accuracy = knn.score(X_test,y_test)
print(f"Training Accuracy of KNN is: {train_accuracy:.2f}")
print(f"Testing Accuracy of KNN is: {test_accuracy:.2f}")
```
